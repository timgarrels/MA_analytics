```python
import matplotlib.pyplot as plt
import networkx as nx

from pmotif_lib.graphlet_representation import GRAPHLET_CLASS_NAME_LOOKUP, graphlet_class_to_graph


fig, axes = plt.subplots(1,len(GRAPHLET_CLASS_NAME_LOOKUP), figsize=(5*len(GRAPHLET_CLASS_NAME_LOOKUP), 5))

for i, (matrix, name) in enumerate(GRAPHLET_CLASS_NAME_LOOKUP.items()):
    ax = axes[i]
    
    nx.draw(graphlet_class_to_graph(matrix), ax=ax)
    ax.set_title(name)
```

```python
fig, axes = plt.subplots(2,1, dpi=1024, figsize=(6, 5))

graphlets_3 = [(k,v) for k, v in GRAPHLET_CLASS_NAME_LOOKUP.items() if len(k.split(" ")[0]) == 3]
for i, (matrix, name) in enumerate(graphlets_3):
        
    ax = axes[i]
    nx.draw(graphlet_class_to_graph(matrix), ax=ax)
    ax.set_title(name)
fig.tight_layout()
```

```python
fig, axes = plt.subplots(2,3, dpi=1024, figsize=(12,5))

axes = [axes[0][0], axes[0][1], axes[0][2], axes[1][0], axes[1][1], axes[1][2]]

graphlets_4 = [(k,v) for k, v in GRAPHLET_CLASS_NAME_LOOKUP.items() if len(k.split(" ")[0]) == 4]

for i, (matrix, name) in enumerate(graphlets_4):
    
    if len(matrix.split(" ")[0]) != 4:
        continue
    
    ax = axes[i]
    pos = None
    if name in ["Fork", "Spoon"]:
        pos = {3: (0, 1), 0: (1, 1), 1: (2, 0), 2: (2, 2)}
    elif "Crossed" in name:
        pos = {0: (0, 0), 3: (1, 0), 2: (0, 1), 1: (1, 1)}
    elif "Square" in name:
        pos = {0: (0, 0), 1: (1, 0), 2: (0, 1), 3: (1, 1)}
        
    
    nx.draw(graphlet_class_to_graph(matrix), ax=ax, pos=pos)
    ax.set_title(name)
fig.tight_layout()
```

```python
fig, axes = plt.subplots(2,4, dpi=1024, figsize=(16,5))
change_names = False
axes = axes.flatten()
axes[1], axes[4] = axes[4], axes[1]

graphlets_4 = [(k,v) for k, v in GRAPHLET_CLASS_NAME_LOOKUP.items() if len(k.split(" ")[0]) == 4]

for j, (matrix, name) in enumerate(graphlets_3):
    ax = axes[j]
    pos = None
    nx.draw(graphlet_class_to_graph(matrix), ax=ax, pos=pos)
    if change_names and name == "Triangle":
        name = "3-clique"
    ax.set_title(name)
    

for i, (matrix, name) in enumerate(graphlets_4):
    
    if len(matrix.split(" ")[0]) != 4:
        continue
    
    ax = axes[i + len(graphlets_3)]
    pos = None
    if name in ["Fork", "Spoon"]:
        pos = {3: (0, 1), 0: (1, 1), 1: (2, 0), 2: (2, 2)}
    elif "Crossed" in name:
        pos = {0: (0, 0), 3: (1, 0), 2: (0, 1), 1: (1, 1)}
    elif "Square" in name:
        pos = {0: (0, 0), 1: (1, 0), 2: (0, 1), 3: (1, 1)}
        
    
    nx.draw(graphlet_class_to_graph(matrix), ax=ax, pos=pos)
    if change_names and name == "Double Crossed Square":
        name = "4-clique"
    ax.set_title(name)
fig.tight_layout()
```
